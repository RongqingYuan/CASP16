```python
file = "/data/data1/conglab/qcong/CASP16/stage2_oligomer_ranksA/step4_best_T0"

score_dict = {}
with open(file, "r") as f:
    for line in f:
        word = line.split()
        score_dict[word[0]] = []
        score_dict[word[0]].append(float(word[2]))
        score_dict[word[0]].append(float(word[3]))
        score_dict[word[0]].append(float(word[4]))
        score_dict[word[0]].append(float(word[5]))

# plot the score, let the bars stand on top of each other
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Extract data from the dictionary
labels = list(score_dict.keys())  # The keys as labels for the x-axis
scores = np.array(list(score_dict.values()))  # Convert values to a NumPy array

# Transpose scores for stacking
scores_transposed = scores.T

# Create the stacked bar chart
x = np.arange(len(labels))  # The x locations for the groups
fig, ax = plt.subplots(figsize=(15, 10))

# Colors for each score component
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
score_names = ['ICS', 'IPS', 'QSbest', 'DockQ']  # Names of the scores

# Plot each score component on top of the previous one
for i in range(scores_transposed.shape[0]):
    if i == 0:
        ax.bar(x, scores_transposed[i], label=score_names[i], color=colors[i])
    else:
        ax.bar(x, scores_transposed[i], bottom=scores_transposed[:i].sum(axis=0), 
               label=score_names[i], color=colors[i])

# Formatting the plot
ax.set_xlabel('Keys')
ax.set_ylabel('Scores')
ax.set_title('Stacked Bar Chart of Scores')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Sort the dictionary by the sum of scores and take the top 25
top_25 = dict(sorted(score_dict.items(), key=lambda item: sum(item[1]), reverse=True)[:25])

# Extract data from the top 25 dictionary
labels = list(top_25.keys())  # The keys as labels for the x-axis
scores = np.array(list(top_25.values()))  # Convert values to a NumPy array

# Transpose scores for stacking
scores_transposed = scores.T

# Create the stacked bar chart
x = np.arange(len(labels))  # The x locations for the groups
fig, ax = plt.subplots(figsize=(12, 8), dpi=300)  # Increased figure size

# Colors for each score component
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
score_names = ['Score1', 'Score2', 'Score3', 'Score4']  # Names of the scores
score_names = ['ICS', 'IPS', 'QSbest', 'DockQ']  # Names of the scores

# Plot each score component on top of the previous one
for i in range(scores_transposed.shape[0]):
    if i == 0:
        ax.bar(x, scores_transposed[i], label=score_names[i], color=colors[i])
    else:
        ax.bar(x, scores_transposed[i], bottom=scores_transposed[:i].sum(axis=0), 
               label=score_names[i], color=colors[i])

# Formatting the plot
ax.set_ylabel('z-score', fontsize=16)
ax.set_title('stage 0, sum z-score, best', fontsize=16, pad = 20)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=16)
ax.set_yticklabels(ax.get_yticks(), fontsize=16)
ax.legend(fontsize=16)

plt.tight_layout()
plt.show()
```
