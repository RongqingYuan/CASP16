```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import sys
import pdb

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

measure = "GDT_TS"
measure = "GDT_HA"
measure = "qs_global"
# model = "first"
model = "best"

# mode = "easy"
# mode = "medium"
# mode = "hard"
mode = "all"

impute_value = -2
os.chdir('/home2/s439906/project/CASP16/oligomer/')

score_path = "./out_v2_np/"
score_file = f"heatmap_custom_{model}_{mode}_impute={impute_value}_equal_weight.csv"
df = pd.read_csv(score_path + score_file, index_col=0)
df
```

```python
TS267 = df.loc['TS267']
TS294 = df.loc['TS294']
# get a separate df where 'TS267', 'TS294' index are removed
indices_to_remove = ['TS267', 'TS294']
df_filtered = df.drop(index=indices_to_remove)
print(df_filtered)
# get the highest value for each column
highest = df_filtered.apply(lambda x: x.nlargest(1).iloc[-1])
print(highest)


diff = TS267 - highest
diff = diff.sort_values(ascending=False)
print(diff)
# plot diff
plt.figure(figsize=(15, 5), dpi=300)
plt.bar(diff.keys(), diff.values)
plt.xticks(rotation=45,ha='right',fontsize=8)
plt.ylabel("z-score")
# draw a line at 0
plt.axhline(0, color='black', lw=1)
plt.title(f"TS267 - highest z-score from other groups" )
plt.show()

diff = TS294 - highest
diff = diff.sort_values(ascending=False)
print(diff)
# plot diff
plt.figure(figsize=(15, 5), dpi=300)
plt.bar(diff.keys(), diff.values)
plt.xticks(rotation=45,ha='right',fontsize=8)
plt.ylabel("z-score")
# draw a line at 0
plt.axhline(0, color='black', lw=1)
plt.title(f"TS294 - highest z-score from other groups" )
plt.show()
```
