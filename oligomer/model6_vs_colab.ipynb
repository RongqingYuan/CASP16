```python
model1_result = "./score_T1_o/dockq-first-all-raw.csv"
model6_result = "./score_T1_o/dockq-sixth-all-raw.csv"

model6 = pd.read_csv(model6_result, index_col=0)
model1 = pd.read_csv(model1_result, index_col=0)
```

```python
# get the TS145 row in model1 as colabfold_baseline
colab_baseline = model1.loc['TS145'].to_frame().T
type(colab_baseline)
colab_baseline
#  concat the colab_baseline to model6
model6_df = pd.concat([model6, colab_baseline], axis=0)
model6_df
```

```python
groups = model6_df.index
# take out every group in the model6_df, get common non-nan values, and run paired t-test
for group in groups:
    group_name = number2group[group]
    if group == 'TS145':
        continue
    baseline_data = model6_df.loc['TS145']
    model6_data = model6_df.loc[group]
    common_non_nan = baseline_data.notna() & model6_data.notna()    
    baseline_common_values = baseline_data[common_non_nan]
    model_6_common_values = model6_data[common_non_nan]
    # if there are more than 1 common values, run paired t-test
    if len(baseline_common_values) > 1:
        t_stat, p_value = ttest_rel(model_6_common_values, baseline_common_values)
        print(f"Group: {group_name}")
        print(len(baseline_common_values))
        print(f"T-statistic: {t_stat}, P-value: {p_value}")
        print("-" * 40)
    else:
        print(f"Group: {group_name} has insufficient data for t-test.")
        print("-" * 40)
```

```python
import pandas as pd
import numpy as np


baseline_values = model6_df.loc['TS145']

normalized_df = pd.DataFrame(index=model6_df.index, columns=model6_df.columns)

# 对每一行与基线行共有非 NaN 值的位置进行除法
for group in model6_df.index:
    group_values = model6_df.loc[group]

    # 找到基线和当前组共有的非 NaN 值
    common_non_nan = baseline_values.notna() & group_values.notna()

    # 在共有非 NaN 值的位置进行除法，其他位置保持为 NaN
    normalized_values = group_values.copy()
    normalized_values[~baseline_values.notna()] = np.nan  # 将基线中为 NaN 的位置设为 NaN
    normalized_values[common_non_nan] = group_values[common_non_nan] - baseline_values[common_non_nan]

    # 存储结果
    normalized_df.loc[group] = normalized_values

# 计算每一行的非 NaN 值的均值
row_means = normalized_df.mean(axis=1)

# 查看结果
print(normalized_df)
mask = normalized_df.isna()
normalized_df_filled = normalized_df.fillna(0)
row_means
# normalized_df.sum(axis=1)
normalized_df
```

```python
sorted_indices = row_means.sort_values(ascending=False).index
sorted_heatmap_data = normalized_df_filled.loc[sorted_indices].reset_index(
    drop=True)
sorted_sum = row_means.loc[sorted_indices].reset_index(drop=True)
sorted_mask = pd.DataFrame(
    mask, index=normalized_df_filled.index).loc[sorted_indices].reset_index(drop=True)
# use mask to mask the data. will be used for heatmap
masked_data = sorted_heatmap_data.copy()
masked_data[sorted_mask] = np.nan
```
