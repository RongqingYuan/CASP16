```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib.gridspec import GridSpec

# 
data = np.random.rand(50, 10)  # 5010
scores = np.random.randint(1, 100, size=50)  # 50

#  DataFrame
df = pd.DataFrame(data, columns=[f'R{i}' for i in range(1, 11)])
score_df = pd.Series(scores)

#  gridspec
fig = plt.figure(figsize=(14, 10))
gs = GridSpec(1, 2, width_ratios=[4, 1], wspace=0.05)

# 
ax0 = fig.add_subplot(gs[0])
sns.heatmap(df, cmap="YlGn", cbar=False, ax=ax0)
ax0.set_yticklabels([f'TS{i}' for i in range(1, 51)], rotation=0)
ax0.set_xticklabels(df.columns, rotation=90)
ax0.set_title("Heatmap")

# 
ax1 = fig.add_subplot(gs[1], sharey=ax0)
ax1.barh(range(len(score_df)), score_df, color='tan')
ax1.set_yticks(range(len(score_df)))
ax1.set_yticklabels([f'TS{i}' for i in range(1, 51)], rotation=0)
ax1.invert_yaxis()  # Y
ax1.set_xlabel("Score")
ax1.set_title("Scores")

# 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap

# 
np.random.seed(0)
data = np.random.rand(10, 10)  # 10x10 
mask = np.random.choice([True, False], size=data.shape, p=[0.3, 0.7])  # 30%

#  NaN
masked_data = np.where(mask, np.nan, data)

#  colormap NaN 
cmap = sns.color_palette("YlGn", as_cmap=True)
cmap.set_bad(color="gray")  #  NaN 

#  mask 
plt.figure(figsize=(8, 6))
sns.heatmap(masked_data, cmap=cmap, cbar=True)
plt.title("Heatmap with Gray Masked Areas for Problematic Data")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap
from matplotlib.gridspec import GridSpec

# 
np.random.seed(0)
data = np.random.rand(50, 10)  # 5010
scores = np.random.randint(1, 100, size=50)  # 50
mask = np.random.choice([True, False], size=data.shape, p=[0.3, 0.7])  # 30%

#  DataFrame  Series
df = pd.DataFrame(data, columns=[f'R{i}' for i in range(1, 11)])
score_df = pd.Series(scores, name="Score")

# 
sorted_indices = score_df.sort_values(ascending=False).index
sorted_df = df.loc[sorted_indices].reset_index(drop=True)
sorted_scores = score_df.loc[sorted_indices].reset_index(drop=True)
sorted_mask = mask[sorted_indices, :]  #  mask 

#  colormap NaN 
cmap = sns.color_palette("YlGn", as_cmap=True)
cmap.set_bad(color="gray")

#  gridspec
fig = plt.figure(figsize=(16, 10))
gs = GridSpec(1, 2, width_ratios=[4, 1], wspace=0.3)

# 
ax0 = fig.add_subplot(gs[0])
sns.heatmap(sorted_df, mask=sorted_mask, cmap=cmap, cbar=True, ax=ax0)
ax0.set_yticklabels([f'TS{i}' for i in sorted_indices + 1], rotation=0)  # 
ax0.set_xticklabels(sorted_df.columns, rotation=90)
ax0.set_title("Heatmap with Gray Masked Areas for Problematic Data")

# 
ax1 = fig.add_subplot(gs[1], sharey=ax0)
ax1.barh(range(len(sorted_scores)), sorted_scores, color='tan')
ax1.set_yticks(range(len(sorted_scores)))
ax1.set_yticklabels([f'TS{i}' for i in sorted_indices + 1], rotation=0)  # 
ax1.invert_yaxis()  # Y
ax1.set_xlabel("Score")
ax1.set_title("Scores")

# 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap
from matplotlib.gridspec import GridSpec

# 
np.random.seed(0)
data = np.random.rand(50, 10)  # 5010
scores = np.random.randint(1, 100, size=50)  # 50
mask = np.random.choice([True, False], size=data.shape, p=[0.3, 0.7])  # 30%

#  DataFrame  Series
df = pd.DataFrame(data, columns=[f'R{i}' for i in range(1, 11)])
score_df = pd.Series(scores, name="Score")

# 
sorted_indices = score_df.sort_values(ascending=False).index
sorted_df = df.loc[sorted_indices].reset_index(drop=True)
sorted_scores = score_df.loc[sorted_indices].reset_index(drop=True)
sorted_mask = mask[sorted_indices, :]  #  mask 

#  mask  True  NaN
masked_data = sorted_df.copy()
masked_data[sorted_mask] = np.nan

#  matplotlib  colormap NaN 
cmap = plt.cm.YlGn
cmap = ListedColormap(cmap(np.linspace(0, 1, 256)))
cmap.set_bad(color='gray')  #  NaN 

#  gridspec
fig = plt.figure(figsize=(16, 10))
gs = GridSpec(1, 2, width_ratios=[4, 1], wspace=0.3)

# 
ax0 = fig.add_subplot(gs[0])
sns.heatmap(masked_data, cmap=cmap, cbar=True, ax=ax0)
ax0.set_yticklabels([f'TS{i}' for i in sorted_indices + 1], rotation=0)  # 
ax0.set_xticklabels(sorted_df.columns, rotation=90)
ax0.set_title("Heatmap with Gray Masked Areas for Problematic Data")

# 
ax1 = fig.add_subplot(gs[1], sharey=ax0)
ax1.barh(range(len(sorted_scores)), sorted_scores, color='tan')
ax1.set_yticks(range(len(sorted_scores)))
ax1.set_yticklabels([f'TS{i}' for i in sorted_indices + 1], rotation=0)  # 
ax1.invert_yaxis()  # Y
ax1.set_xlabel("Score")
ax1.set_title("Scores")

# 
plt.tight_layout()
plt.show()
```
