```python
import os
processed_path ="/home2/s439906/project/CASP16/monomer/monomer_data_Sep_15_EU/processed"
csv_s = [file for file in os.listdir(processed_path) if file.endswith('.csv')]

qse_df = pd.DataFrame()
for csv in csv_s:
    csv_path = os.path.join(processed_path, csv)
    df = pd.read_csv(csv_path, index_col=0)
    # get QSE column
    QSE = pd.DataFrame(df['QSE'])
    QSE.index = QSE.index.str.extract(
            r'(T\w+)TS(\w+)_(\w+)-(D\w+)').apply(lambda x: (f"{x[0]}-{x[3]}", f"TS{x[1]}", x[2]), axis=1)
    QSE.index = pd.MultiIndex.from_tuples(
            QSE.index, names=['target', 'group', 'submission_id'])
        # # get all data with submission_id == 6
        # data_tmp = data_tmp.loc[(slice(None), slice(None), "6"), :]
        # drop all data with submission_id == 6
    # QSE = QSE.loc[(slice(None), slice(None), ["1"]), :]
    QSE = QSE.loc[(slice(None), slice(None), [ "1", "2", "3", "4", "5"]), :]
    grouped = QSE.groupby(["group"])
    grouped = pd.DataFrame(grouped["QSE"].max())
    # add to qse_df
    if qse_df.empty:
        qse_df = grouped
    else:
        qse_df = pd.concat([qse_df, grouped], axis=1)
sum = qse_df.sum(axis=1)
qse_df['sum'] = sum
qse_df = qse_df.sort_values('sum', ascending=False)
qse_df.head(15)
```
