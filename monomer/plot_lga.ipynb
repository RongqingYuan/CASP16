```python
# plot the lga results I computed as a reference

best_model_result_dir = "./model_lga_job/best_job/TMP/"
first_model_result_dir = "./model_lga_job/first_job/TMP/"
first_of_best_model_result_dir = "./model_lga_job/first_of_best_job/TMP/"


best_model_lga_result = {}
first_model_lga_result = {}
first_of_best_model_lga_result = {}

for file in os.listdir(best_model_result_dir):
    target_name = file.split("TS")[0]
    domain_name = file.split("-")[1].split(".")[0]
    EU_name = target_name + "-" + domain_name
    if file.endswith(".lga"):
        with open(os.path.join(best_model_result_dir, file), 'r') as f:
            for line in f:
                if line.startswith("SUMMARY(GDT)"):
                    words = line.split()
                    GDT_HA = float(words[8])
                    if GDT_HA < 35:
                        print("{} is not good enough, please check the file".format(EU_name))
                    if EU_name in best_model_lga_result:
                        print("{} already exists, please check the file".format(EU_name))
                    else:
                        best_model_lga_result[EU_name] = GDT_HA


for file in os.listdir(first_model_result_dir):
    target_name = file.split("TS")[0]
    domain_name = file.split("-")[1].split(".")[0]
    EU_name = target_name + "-" + domain_name
    if file.endswith(".lga"):
        with open(os.path.join(first_model_result_dir, file), 'r') as f:
            for line in f:
                if line.startswith("SUMMARY(GDT)"):
                    words = line.split()
                    GDT_HA = float(words[8])
                    if GDT_HA < 35:
                        print("{} is not good enough, please check the file".format(EU_name))
                    if EU_name in first_model_lga_result:
                        print("{} already exists, please check the file".format(EU_name))
                    else:
                        first_model_lga_result[EU_name] = GDT_HA
for file in os.listdir(first_of_best_model_result_dir):
    target_name = file.split("TS")[0]
    domain_name = file.split("-")[1].split(".")[0]
    EU_name = target_name + "-" + domain_name
    if file.endswith(".lga"):
        with open(os.path.join(first_of_best_model_result_dir, file), 'r') as f:
            for line in f:
                if line.startswith("SUMMARY(GDT)"):
                    words = line.split()
                    GDT_HA = float(words[8])
                    if GDT_HA < 35:
                        print("{} is not good enough, please check the file".format(EU_name))
                    if EU_name in first_of_best_model_lga_result:
                        print("{} already exists, please check the file".format(EU_name))
                    else:
                        first_of_best_model_lga_result[EU_name] = GDT_HA
```

```python
massivefold_root_dir = "./lga_output/"
EUs = [dir for dir in os.listdir(massivefold_root_dir) if os.path.isdir(os.path.join(massivefold_root_dir, dir))]
EUs.sort()
massivefold_lga_result = {}
for EU in EUs:
    files_count = 0
    best_GDT_HA = 0
    best_file_name = None
    EU_dir = os.path.join(massivefold_root_dir, EU)
    sub_dirs = [dir for dir in os.listdir(EU_dir) if os.path.isdir(os.path.join(EU_dir, dir))]
    for sub_dir in sub_dirs:
        sub_dir_path = os.path.join(EU_dir, sub_dir)
        for file in os.listdir(sub_dir_path):
            if file.endswith(".lga"):
                with open(os.path.join(sub_dir_path, file), 'r') as f:
                    for line in f:
                        if line.startswith("SUMMARY(GDT)"):
                            words = line.split()
                            GDT_HA = float(words[8])
                            if GDT_HA > best_GDT_HA:
                                best_GDT_HA = GDT_HA
                                best_file_name = sub_dir_path.split("/")[-1] + "/" + file.split(".")[0]
            files_count += 1
    massivefold_lga_result[EU] = [best_GDT_HA, best_file_name, files_count]
    print(f"{EU} has {files_count} models, the best model is {best_file_name} with GDT_HA {best_GDT_HA}")

```

```python
best_model_lga_result = dict(sorted(best_model_lga_result.items(), key=lambda x: x[1], reverse=True))
first_model_lga_result = dict(sorted(first_model_lga_result.items(), key=lambda item: best_model_lga_result[item[0]], reverse=True))
first_of_best_model_lga_result = dict(sorted(first_of_best_model_lga_result.items(), key=lambda item: best_model_lga_result[item[0]], reverse=True))
massivefold_lga_score = dict(sorted(massivefold_lga_score.items(), key=lambda item: best_model_lga_result[item[0]], reverse=True))

targets = list(best_model_lga_result.keys())
best_model_values = list(best_model_lga_result.values())
first_model_values = list(first_model_lga_result.values())
first_of_best_model_values = list(first_of_best_model_lga_result.values())
massivefold_values = list(massivefold_lga_score.values())

# Function to calculate moving average
def moving_average(data, window_size):
    if window_size > len(data):
        window_size = len(data)
    
    # Create a pandas Series and use rolling method
    return pd.Series(data).rolling(window=window_size, center=True).mean().fillna(method='bfill').fillna(method='ffill').tolist()

# Choose window size for moving average (adjust as needed)
window_size = 1  # Smaller for less smoothing, larger for more smoothing

best_model_ma = moving_average(best_model_values, window_size)
first_model_ma = moving_average(first_model_values, window_size)
first_of_best_model_ma = moving_average(first_of_best_model_values, window_size)
massivefold_ma = moving_average(massivefold_values, window_size)
plt.figure(figsize=(16, 6), dpi=300)
plt.scatter(targets, best_model_values, marker='o', label='Best Model', alpha=0.5)
plt.scatter(targets, first_model_values, marker='s', label='Best First Model', alpha=0.5)
plt.scatter(targets, first_of_best_model_values, marker='o', label='First Model of Best Group', alpha=0.5)
plt.scatter(targets, massivefold_values, marker='x', label='MassiveFold', alpha=0.5)

# plt.plot(targets, best_model_ma,'-', label=f'Best Model Moving Average ({window_size})',linewidth=3)
# plt.plot(targets, first_model_ma,'-', label=f'First Model Moving Average ({window_size})',linewidth=3)
# plt.plot(targets, first_of_best_model_ma,'-', label=f'First of Best Group Moving Average ({window_size})', linewidth=3)
# plt.plot(targets, massivefold_ma,'-', label=f'MassiveFold Moving Average ({window_size})', linewidth=3)

plt.plot(targets, best_model_ma,'-', label='Best Model',linewidth=3)
plt.plot(targets, first_model_ma,'-', label='Best First Model',linewidth=3)
plt.plot(targets, first_of_best_model_ma,'-', label='First Model of Best Group', linewidth=3)
plt.plot(targets, massivefold_ma,'-', label='MassiveFold', linewidth=3)

plt.xlabel('Targets', fontsize=16)
plt.xticks(rotation=90,fontsize=16)
plt.ylabel('GDT_HA', fontsize=16)
plt.yticks(fontsize=16)
# plt.title('GDT_HA of Best Model and Moving Average')
plt.ylim(0,100)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=16)
plt.tight_layout()
plt.show()
```
