```python
import os
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
os.chdir('/home2/s439906/project/CASP16/monomer/')


measure = "GDT_TS"
measure = "GDT_HA"

# model = "first"
model = "best"

# mode = "easy"
# mode = "medium"
# mode = "hard"
mode = "all"

impute_value = -2


score_path = "./heatmap/"
score_file = f"heatmap_CASP16_{model}_{mode}_impute={impute_value}_equal_weight.csv"
df = pd.read_csv(score_path + score_file, index_col=0)



print(df)
print(df.max())
```

```python
# count how many times each group get the highest score in each column of T1_data
# Iterate through each column and add to the count for rows with the maximum value in that column
highest_count = pd.Series(0, index=df.index)
for col in df.columns:
    max_value = df[col].max()  # Find the maximum value in the column
    highest_count += (df[col] == max_value)  # Add 1 to rows with the max value

# print(highest_count)
# sort the count in descending order
highest_count = highest_count.sort_values(ascending=False)
highest_count=highest_count.to_dict()
highest_count
```

```python
# plot highest_count
plt.figure(figsize=(15, 5), dpi=300)
plt.bar(highest_count.keys(), highest_count.values())
plt.xticks(rotation=90,fontsize=8)
plt.title(f"Number of times each group get the highest score for {measure}")
plt.show()
```
