```python
import pandas as pd
```

```python

model1_result = "/home2/s439906/project/CASP16/monomer/7_model_6/group_by_target_EU/GDT_HA-1-all-raw.csv"
model6_result = "/home2/s439906/project/CASP16/monomer/7_model_6/group_by_target_EU/GDT_HA-6-all-raw.csv"

model1 = pd.read_csv(model1_result, index_col=0)
model6 = pd.read_csv(model6_result, index_col=0)
```

```python
# get the comm rows in two dataframes
common_rows = model1.index.intersection(model6.index)
# get the df with common rows
model1_common = model1.loc[common_rows]
model6_common = model6.loc[common_rows]
model6_common
model1_common
```

```python
# element-wise division of two dataframes
division = model6_common.div(model1_common)
mask = division.isna()
division
```

```python
# sum of the division
# fill division by 0
division_fill = division.fillna(0)
sum_division = division.sum(axis=1)
# only average on non-NaN values
```

```python

# divide the sum by the number of non-NaN values for each row
num_non_nan = division.count(axis=1)
num_non_nan

average_division = sum_division / num_non_nan
average_division
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
from matplotlib.colors import ListedColormap
import numpy as np

# sum = division_fill.sum(axis=1)
sum = average_division
sorted_indices = sum.sort_values(ascending=True).index
sorted_heatmap_data = division_fill.loc[sorted_indices].reset_index(
    drop=True)
sorted_sum = sum.loc[sorted_indices].reset_index(drop=True)
sorted_mask = pd.DataFrame(
    mask, index=division_fill.index).loc[sorted_indices].reset_index(drop=True)
# use mask to mask the data. will be used for heatmap
masked_data = sorted_heatmap_data.copy()
masked_data[sorted_mask] = np.nan
# set up the colormap
cmap = plt.cm.YlGn
cmap = ListedColormap(cmap(np.linspace(0, 1, 256)))
cmap.set_bad(color='gray')  # set the masked area to gray
# set up the figure and gridspec
fig = plt.figure(figsize=(32, 4), dpi=300)
gs = GridSpec(1, 2, width_ratios=[4, 1], wspace=0.04)
# plot the heatmap
ax0 = fig.add_subplot(gs[0])
sns.heatmap(masked_data, cmap=cmap, cbar=True, ax=ax0)
ax0.set_yticklabels(
    [f'{i}' for i in sorted_indices], rotation=0)  # use the same order as the row sum
ax0.set_xticklabels(sorted_heatmap_data.columns, rotation=90)
# set x tick font size
ax0.tick_params(axis='x', labelsize=16)
# set y tick font size
ax0.tick_params(axis='y', labelsize=16)
# set the font size of the colorbar
cbar = ax0.collections[0].colorbar
cbar.ax.tick_params(labelsize=16)

ax0.set_title(
    "model 6 (GDT_HA) / model 1 (GDT_HA)", fontsize=20)
# plot the row sum
ax1 = fig.add_subplot(gs[1], sharey=ax0)
y_pos = range(len(sorted_sum))
y_pos = [i+0.5 for i in y_pos]  # change the position of the bars
ax1.barh(y_pos, sorted_sum, color='tan')
# ax1.margins(y=0.5)
ax1.set_yticks(range(len(sorted_sum)))
ax1.set_yticklabels(
    [f'{i}' for i in sorted_indices], rotation=0)  # use the same order as the heatmap
# ax1.spines['bottom'].set_position(('outward', 10))  # 将 x 轴向下移动 10 点
# ymin, ymax = ax1.get_ylim()  # 获取当前的 y 轴范围
# ax1.set_ylim(ymin - 1, ymax-1)  # 为最底部条形预留空间
# set x tick font size
ax1.tick_params(axis='x', labelsize=16)
# set y tick font size
ax1.tick_params(axis='y', labelsize=16)
ax1.invert_yaxis()  # flip the y-axis
ax1.set_xlabel("Sum", fontsize=16)
ax1.set_title("Group average", fontsize=20)
# draw a line at x=1 
ax1.axvline(x=1, color='r', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
from matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd

# 示例数据
np.random.seed(0)
division_fill = pd.DataFrame(np.random.rand(20, 20), columns=[f"Col_{i}" for i in range(20)])
mask = np.random.choice([True, False], size=division_fill.shape)
average_division = division_fill.sum(axis=1)

# 计算和排序
sum = average_division
sorted_indices = sum.sort_values(ascending=True).index
sorted_heatmap_data = division_fill.loc[sorted_indices].reset_index(drop=True)
sorted_sum = sum.loc[sorted_indices].reset_index(drop=True)
sorted_mask = pd.DataFrame(mask, index=division_fill.index).loc[sorted_indices].reset_index(drop=True)

# 使用掩码遮蔽数据
masked_data = sorted_heatmap_data.copy()
masked_data[sorted_mask] = np.nan

# 随机生成行注解标签
row_annotation = pd.DataFrame({
    'Category': np.random.choice(['Group 1', 'Group 2', 'Group 3'], size=len(sorted_heatmap_data))
}, index=sorted_heatmap_data.index)

# 设置颜色映射
cmap = plt.cm.YlGn
cmap = ListedColormap(cmap(np.linspace(0, 1, 256)))
cmap.set_bad(color='gray')  # 设置遮蔽区域颜色为灰色

# 设置图形和GridSpec
fig = plt.figure(figsize=(32, 4), dpi=300)
gs = GridSpec(1, 3, width_ratios=[0.2, 4, 1], wspace=0.04)

# 添加行注解的颜色条
ax_row_colors = fig.add_subplot(gs[0])
row_colors = row_annotation['Category'].map({'Group 1': 'red', 'Group 2': 'blue', 'Group 3': 'green'})
for idx, color in enumerate(row_colors):
    ax_row_colors.add_patch(plt.Rectangle((0, idx), 1, 1, color=color))
ax_row_colors.set_xlim(0, 1)
ax_row_colors.set_ylim(len(row_colors), 0)
ax_row_colors.axis("off")  # 隐藏坐标轴

# 绘制热力图
ax0 = fig.add_subplot(gs[1])
sns.heatmap(masked_data, cmap=cmap, cbar=True, ax=ax0)
ax0.set_yticklabels([f'{i}' for i in sorted_indices], rotation=0)
ax0.set_xticklabels(sorted_heatmap_data.columns, rotation=90)
ax0.tick_params(axis='x', labelsize=16)
ax0.tick_params(axis='y', labelsize=16)

# 设置 colorbar 字体大小
cbar = ax0.collections[0].colorbar
cbar.ax.tick_params(labelsize=16)
ax0.set_title("model 6 (GDT_HA) / model 1 (GDT_HA)", fontsize=20)

# 绘制行和的条形图
ax1 = fig.add_subplot(gs[2], sharey=ax0)
y_pos = range(len(sorted_sum))
y_pos = [i+0.5 for i in y_pos]
ax1.barh(y_pos, sorted_sum, color='tan')
ax1.set_yticks(range(len(sorted_sum)))
ax1.set_yticklabels([f'{i}' for i in sorted_indices], rotation=0)
ax1.tick_params(axis='x', labelsize=16)
ax1.tick_params(axis='y', labelsize=16)
ax1.invert_yaxis()
ax1.set_xlabel("Sum", fontsize=16)
ax1.set_title("Group average", fontsize=20)
ax1.axvline(x=1, color='r', linestyle='--')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
from matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd

# 示例数据
np.random.seed(0)
division_fill = pd.DataFrame(np.random.rand(20, 20), columns=[f"Col_{i}" for i in range(20)])
mask = np.random.choice([True, False], size=division_fill.shape)
average_division = division_fill.sum(axis=1)

# 计算和排序
sum = average_division
sorted_indices = sum.sort_values(ascending=True).index
sorted_heatmap_data = division_fill.loc[sorted_indices].reset_index(drop=True)
sorted_sum = sum.loc[sorted_indices].reset_index(drop=True)
sorted_mask = pd.DataFrame(mask, index=division_fill.index).loc[sorted_indices].reset_index(drop=True)

# 使用掩码遮蔽数据
masked_data = sorted_heatmap_data.copy()
masked_data[sorted_mask] = np.nan

# 随机生成三种不同的行注解
row_annotation = pd.DataFrame({
    'Category 1': np.random.choice(['Group A', 'Group B', 'Group C'], size=len(sorted_heatmap_data)),
    'Category 2': np.random.choice(['Type X', 'Type Y'], size=len(sorted_heatmap_data)),
    'Category 3': np.random.choice(['High', 'Medium', 'Low'], size=len(sorted_heatmap_data))
}, index=sorted_heatmap_data.index)

# 定义每个注解类别的颜色映射
colors_1 = {'Group A': 'red', 'Group B': 'blue', 'Group C': 'green'}
colors_2 = {'Type X': 'purple', 'Type Y': 'orange'}
colors_3 = {'High': 'darkgreen', 'Medium': 'yellow', 'Low': 'lightgrey'}

# 设置颜色映射
cmap = plt.cm.YlGn
cmap = ListedColormap(cmap(np.linspace(0, 1, 256)))
cmap.set_bad(color='gray')  # 设置遮蔽区域颜色为灰色

# 设置图形和GridSpec
fig = plt.figure(figsize=(34, 4), dpi=300)
gs = GridSpec(1, 5, width_ratios=[0.2, 0.2, 0.2, 4, 1], wspace=0.04)

# 添加每个注解的颜色条
# Category 1
ax_cat1 = fig.add_subplot(gs[0])
row_colors_1 = row_annotation['Category 1'].map(colors_1)
for idx, color in enumerate(row_colors_1):
    ax_cat1.add_patch(plt.Rectangle((0, idx), 1, 1, color=color))
ax_cat1.set_xlim(0, 1)
ax_cat1.set_ylim(len(row_colors_1), 0)
ax_cat1.axis("off")  # 隐藏坐标轴
ax_cat1.set_title("Category 1", fontsize=10, rotation=90, ha='right', va='center')

# Category 2
ax_cat2 = fig.add_subplot(gs[1])
row_colors_2 = row_annotation['Category 2'].map(colors_2)
for idx, color in enumerate(row_colors_2):
    ax_cat2.add_patch(plt.Rectangle((0, idx), 1, 1, color=color))
ax_cat2.set_xlim(0, 1)
ax_cat2.set_ylim(len(row_colors_2), 0)
ax_cat2.axis("off")  # 隐藏坐标轴
ax_cat2.set_title("Category 2", fontsize=10, rotation=90, ha='right', va='center')

# Category 3
ax_cat3 = fig.add_subplot(gs[2])
row_colors_3 = row_annotation['Category 3'].map(colors_3)
for idx, color in enumerate(row_colors_3):
    ax_cat3.add_patch(plt.Rectangle((0, idx), 1, 1, color=color))
ax_cat3.set_xlim(0, 1)
ax_cat3.set_ylim(len(row_colors_3), 0)
ax_cat3.axis("off")  # 隐藏坐标轴
ax_cat3.set_title("Category 3", fontsize=10, rotation=90, ha='right', va='center')

# 绘制热力图
ax0 = fig.add_subplot(gs[3])
sns.heatmap(masked_data, cmap=cmap, cbar=True, ax=ax0)
ax0.set_yticklabels([f'{i}' for i in sorted_indices], rotation=0)
ax0.set_xticklabels(sorted_heatmap_data.columns, rotation=90)
ax0.tick_params(axis='x', labelsize=16)
ax0.tick_params(axis='y', labelsize=16)

# 设置 colorbar 字体大小
cbar = ax0.collections[0].colorbar
cbar.ax.tick_params(labelsize=16)
ax0.set_title("model 6 (GDT_HA) / model 1 (GDT_HA)", fontsize=20)

# 绘制行和的条形图
ax1 = fig.add_subplot(gs[4], sharey=ax0)
y_pos = range(len(sorted_sum))
y_pos = [i+0.5 for i in y_pos]
ax1.barh(y_pos, sorted_sum, color='tan')
ax1.set_yticks(range(len(sorted_sum)))
ax1.set_yticklabels([f'{i}' for i in sorted_indices], rotation=0)
ax1.tick_params(axis='x', labelsize=16)
ax1.tick_params(axis='y', labelsize=16)
ax1.invert_yaxis()
ax1.set_xlabel("Sum", fontsize=16)
ax1.set_title("Group average", fontsize=20)
ax1.axvline(x=1, color='r', linestyle='--')

plt.tight_layout()
plt.show()
```
