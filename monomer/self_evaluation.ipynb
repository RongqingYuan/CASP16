```python
measure = 'GDT_HA'
best_model_file = f'./score_raw_final_T1/{measure}-best-all-raw.csv'
first_model_file = f'./score_raw_final_T1/{measure}-first-all-raw.csv'
best_df = pd.read_csv(best_model_file, index_col=0)
first_df = pd.read_csv(first_model_file, index_col=0)
targets = best_df.columns
best_dict = {}
first_dict = {}
first_of_best_dict = {}

for target in targets:
    best_target_column = best_df[target]
    first_target_column = first_df[target]
    best_target_column = best_target_column.sort_values(ascending=False)
    first_target_column = first_target_column.sort_values(ascending=False)
    # check if there is tied highest score
    highest_best_score = best_target_column.iloc[0]
    highest_first_score = first_target_column.iloc[0]
    tied_best_models = best_target_column[best_target_column == highest_best_score]
    
    if len(tied_best_models) > 1:
        print(f"Target {target} has {len(tied_best_models)} models tied for highest score ({highest_best_score:.2f}):")
        # get each first model of the tied best models
        first_scores_of_tied = {}
        for group in tied_best_models.index:
            first_scores_of_tied[group] = first_target_column.loc[group]
            print(f"  - {group}: first model score = {first_scores_of_tied[group]:.2f}")
        
        # Select the model with the highest first model score as tie-breaker.
        # if there is still tied, then just randomly select one.
        # it is not about ranking so it is ok.
        best_group = max(first_scores_of_tied.items(), key=lambda x: x[1])[0]
        print(f"  Selecting {best_group} as it has the highest first model score ({first_scores_of_tied[best_group]:.2f})")
        best_dict[target] = highest_best_score
        first_dict[target] = highest_first_score
        first_of_best_dict[target] = first_scores_of_tied[best_group]
    else:
        best_group = best_target_column.index[0]
        print(f"Target {target} has no tied models. The best model is {best_group} with score {highest_best_score:.2f}.")
        best_dict[target] = highest_best_score
        first_dict[target] = highest_first_score
        first_of_best_dict[target] = first_target_column.loc[best_group]

print(best_dict)
print(first_dict)
print(first_of_best_dict)

# sort the best_dict by the value, and sort the first_dict and first_of_best_dict by the same order
sorted_best_dict = dict(sorted(best_dict.items(), key=lambda item: item[1], reverse=True))
sorted_first_dict = dict(sorted(first_dict.items(), key=lambda item: sorted_best_dict[item[0]], reverse=True))
sorted_first_of_best_dict = dict(sorted(first_of_best_dict.items(), key=lambda item: sorted_best_dict[item[0]], reverse=True))

print(sorted_best_dict)
print(sorted_first_dict)
print(sorted_first_of_best_dict)



# Create figure and axis
plt.figure(figsize=(14, 8))

# Get targets and values
targets = list(sorted_best_dict.keys())
best_values = list(sorted_best_dict.values())
first_values = list(sorted_first_dict.values())
first_of_best_values = list(sorted_first_of_best_dict.values())

# create a line plot to better visualize the trend
plt.figure(figsize=(14, 8))

plt.plot(targets, best_values, 'o-', label='Best Model', linewidth=2, markersize=8)
plt.plot(targets, first_values, 's-', label='First Model', linewidth=2, markersize=8)
plt.plot(targets, first_of_best_values, '^-', label='First of Best Model', linewidth=2, markersize=8)

plt.xlabel('Targets', fontsize=12)
plt.ylabel(f'{measure} Score', fontsize=12)
plt.title(f'Trend of {measure} Scores Across Targets', fontsize=14)
plt.xticks(rotation=90)
plt.ylim(0, 100)  # Assuming GDT_HA scores range from 0 to 100

plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
