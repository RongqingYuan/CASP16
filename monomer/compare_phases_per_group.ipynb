```python
measure = "GDT_TS"
measure = "GDT_HA"

model = "first"
# model = "best"

# mode = "easy"
# mode = "medium"
# mode = "hard"
mode = "all"

impute_value = -2

T1_score_path = "./score_T1/"
T1_score_file = f"{measure}-{model}-{mode}-raw.csv"
T1_data = pd.read_csv(T1_score_path + T1_score_file, index_col=0)
print(T1_data.shape)
T0_score_path = "./score_T0/"
T0_score_file = f"{measure}-{model}-{mode}-raw.csv"
T0_data = pd.read_csv(T0_score_path + T0_score_file, index_col=0)
print(T0_data.shape)
T2_score_path = "./score_T2/"
T2_score_file = f"{measure}-{model}-{mode}-raw.csv"
T2_data = pd.read_csv(T2_score_path + T2_score_file, index_col=0)
print(T2_data.shape)
# T2_data


# print(T0_data.shape)
# print(T1_data.shape)
# print(T2_data.shape)
out_path = "./CASP16_figures/"
if not os.path.exists(out_path):
    os.makedirs(out_path)
T0_data = T0_data.T
T1_data = T1_data.T
T2_data = T2_data.T

stages = ["T0", "T1"]
stages = ["T1", "T2"]
top_n = 10


# T1 -T0
T0_data.index = T0_data.index.str.extract(r'T0([-\w]+)')[0]
# there is a -D*. In order to match the "-" in the regex, we need to add "-" in the regex
T1_data.index = T1_data.index.str.extract(r'T1([-\w]+)')[0]
# then remove the "-all" in T1_data.index
# T1_data.index = T1_data.index.str.replace("-all", "")
common_index = T0_data.index.intersection(T1_data.index)
df0_aligned = T0_data.loc[common_index]
df1_aligned = T1_data.loc[common_index]
groups = df1_aligned.columns

top_groups =["TS022",
"TS456",
"TS052",
"TS110",
"TS051",
"TS345",
"TS148",
"TS462",
"TS241",
"TS019",
"TS147",
"TS304",
"TS294",
"TS312",
"TS208",
"TS264",
"TS015",
"TS267",
"TS221",
"TS419",
"TS293",
"TS314",
"TS164",
"TS235",
"TS079",
"TS163",
"TS284",
"TS301",]
groups = [group for group in groups if group in top_groups]
diff_valid_dict = {}
diff_valid_numbers = {}
for group in groups:
    # pdb.set_trace()
    try:
        df0_aligned_target = df0_aligned[group]
    except KeyError:
        print(f"{group} not in T0")
        continue
    df1_aligned_target = df1_aligned[group]
    # take the intersection of the two dataframes where both are not null
    # 布尔索引，选择两个列都不为 NaN 的行
    valid_rows = df0_aligned_target.notna() & df1_aligned_target.notna()
    # 取出交集部分
    df0_aligned_valid = df0_aligned_target[valid_rows]
    df1_aligned_valid = df1_aligned_target[valid_rows]
    shape_0 = df0_aligned_valid.shape
    shape_1 = df1_aligned_valid.shape
    assert shape_0 == shape_1
    diff = df1_aligned_valid - df0_aligned_valid
    diff_mean = diff.mean()
    if int(shape_0[0]) > 10:
        diff_valid_dict[group] = diff_mean
        diff_valid_numbers[group] = int(shape_0[0])
# sort the dictionary by the mean of the difference
diff_valid_dict = dict(sorted(diff_valid_dict.items(), key=lambda item: item[1], reverse=True))
diff_valid_dict
group_diff_valid_dict = {}
for group in diff_valid_dict.keys():
    group_diff_valid_dict[number2group[group]] = diff_valid_dict[group]
group_diff_valid_numbers = {}
for group in diff_valid_numbers.keys():
    group_diff_valid_numbers[number2group[group]] = diff_valid_numbers[group]

# plot. add the value of diff_valid_numbers on each bar
plt.figure(figsize=(8, 5), dpi=300)
plt.bar(group_diff_valid_dict.keys(), group_diff_valid_dict.values())
for i, (key, value) in enumerate(group_diff_valid_dict.items()):
    if group_diff_valid_dict[key] > 0:
        plt.text(i, value, group_diff_valid_numbers[key], ha='center')
    else:
        # move the text to the bottom
        plt.text(i, value, group_diff_valid_numbers[key], ha='center', va='top')
     
plt.xticks(fontsize=12, rotation=90)
plt.title(f"mean {measure} improvement from T0 to T1, using {model} model", pad=20)
plt.ylabel(f"mean {measure} improvement")
max_value = max(group_diff_valid_dict.values())
min_value = min(group_diff_valid_dict.values())
plt.ylim(min_value-1, max_value+1)
plt.tight_layout()
plt.show()
```
