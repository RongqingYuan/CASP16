```python
import numpy as np
from Bio.PDB import PDBParser
from Bio.PDB.NeighborSearch import NeighborSearch
# find the residues near the hetero atoms
def find_residues_near_hetero(pdb_file, hetero_residue_names, distance_cutoff=5.0):
    """
    Find all protein residues within a specified distance of hetero atoms.
    
    Parameters:
    -----------
    pdb_file : str
        Path to the PDB file
    hetero_residue_names : list
        List of hetero residue names to search for (e.g., ['ATP', 'MG', 'NAG'])
    distance_cutoff : float, optional
        Distance cutoff in Angstroms (default: 5.0)
    
    Returns:
    --------
    dict
        Dictionary with hetero residues as keys and lists of nearby protein residues as values
    """
    # Parse the PDB file
    parser = PDBParser(QUIET=True)
    structure = parser.get_structure('structure', pdb_file)
    
    # Initialize results dictionary
    nearby_residues = {}
    
    # Get all atoms in the structure
    all_atoms = [atom for atom in structure.get_atoms()]
    
    # Create a neighbor search object
    ns = NeighborSearch(all_atoms)
    
    # Find hetero residues of interest
    for model in structure:
        for chain in model:
            for residue in chain:
                res_id = residue.get_id()
                # Check if it's a hetero residue we're interested in
                if res_id[0].strip() and residue.get_resname() in hetero_residue_names:
                    hetero_key = f"{residue.get_resname()} (Chain {chain.id}, Res #{res_id[1]})"
                    nearby_residues[hetero_key] = []
                    
                    # Get all atoms in this hetero residue
                    hetero_atoms = list(residue.get_atoms())
                    
                    # Find all atoms within distance_cutoff of any hetero atom
                    nearby_atoms = set()
                    for atom in hetero_atoms:
                        nearby_atoms.update(ns.search(atom.get_coord(), distance_cutoff))
                    
                    # Get the residues of these atoms (excluding the hetero residue itself)
                    for atom in nearby_atoms:
                        res = atom.get_parent()
                        res_id = res.get_id()
                        # Only include protein residues (not hetero, not water)
                        if res_id[0] == " " and res != residue:
                            res_info = f"{res.get_resname()} (Chain {res.get_parent().id}, Res #{res_id[1]})"
                            if res_info not in nearby_residues[hetero_key]:
                                nearby_residues[hetero_key].append(res_info)
    
    return nearby_residues


# H1208: Zn
# H1220: Zn
# H1227: Zn
# H1230: Zn
# H1244: Ca
# T1231: Zn
# T1269: Cu, Ca, NAG (NAG attached to a glycoprotein)
# T1292: EDO, Fe (isocyanide hydratase-like iron binding protein. EDO is solvent)
# T1299: Zn

# T1237: C2E and SAM (C2E is additive. I cannot find what protein it is but SAM could a methyl donor. BLAST indicates it is a methyl transferase)
# T1243: 16P, DXE, EDO, MLY, MLZ, SO4 (Cannot find the description of the protein. Heteroatoms looks like solvents or PTMs. UniProt indicates it is a synthase, so only 16P is biologically relevant)
# T1246: GSH, Cl  (Cl is solvent)
# T1278: BG6, EDO, PEG, PG4, PGE (only BG6 is biologically relevant. it should be a dehydrogenase)

CASP16_substrate_dict = {
    "T1237": ["SAM"],
    "T1243": ["16P"],
    "T1246": ["GSH"],
    "T1278": ["BG6"],
}
CASP16_molecule_nearby_residues = {}
for pdb_file in pdb_files:
    pdb_name = os.path.basename(pdb_file)
    submit_target = pdb_name[:5]
    if submit_target in CASP16_substrate_dict:
        results = find_residues_near_hetero(pdb_file, CASP16_substrate_dict[submit_target])
        print(pdb_name, ":",results)
        CASP16_molecule_nearby_residues[pdb_name] = results

CASP16_metal_dict = {
    "H1208": ["ZN"],
    "H1220": ["ZN"],
    "H1227": ["ZN"],
    "H1230": ["ZN"],
    "H1244": ["CA"],
    "T1231": ["ZN"],
    "T1269": ["CU", "CA"],
    "T1292": ["FE"],
    "T1299": ["ZN"],
}
CASP16_metal_nearby_residues = {}
for pdb_file in pdb_files:
    pdb_name = os.path.basename(pdb_file)
    submit_target = pdb_name[:5]
    if submit_target in CASP16_metal_dict:
        results = find_residues_near_hetero(pdb_file, CASP16_metal_dict[submit_target])
        print(pdb_name, ":",results)
        CASP16_metal_nearby_residues[pdb_name] = results
```

```python
import numpy as np
from Bio.PDB import PDBParser
from Bio.PDB.NeighborSearch import NeighborSearch
# find the residues near the hetero atoms
def find_residues_near_hetero(pdb_file, hetero_residue_names, distance_cutoff=5.0):
    """
    Find all protein residues within a specified distance of hetero atoms.
    
    Parameters:
    -----------
    pdb_file : str
        Path to the PDB file
    hetero_residue_names : list
        List of hetero residue names to search for (e.g., ['ATP', 'MG', 'NAG'])
    distance_cutoff : float, optional
        Distance cutoff in Angstroms (default: 5.0)
    
    Returns:
    --------
    dict
        Dictionary with hetero residues as keys and lists of nearby protein residues as values
    """
    # Parse the PDB file
    parser = PDBParser(QUIET=True)
    structure = parser.get_structure('structure', pdb_file)
    
    # Initialize results dictionary
    nearby_residues = {}
    
    # Get all atoms in the structure
    all_atoms = [atom for atom in structure.get_atoms()]
    
    # Create a neighbor search object
    ns = NeighborSearch(all_atoms)
    
    # Find hetero residues of interest
    for model in structure:
        for chain in model:
            for residue in chain:
                res_id = residue.get_id()
                # Check if it's a hetero residue we're interested in
                if res_id[0].strip() and residue.get_resname() in hetero_residue_names:
                    hetero_key = f"{residue.get_resname()} (Chain {chain.id}, Res #{res_id[1]})"
                    nearby_residues[hetero_key] = []
                    
                    # Get all atoms in this hetero residue
                    hetero_atoms = list(residue.get_atoms())
                    
                    # Find all atoms within distance_cutoff of any hetero atom
                    nearby_atoms = set()
                    for atom in hetero_atoms:
                        nearby_atoms.update(ns.search(atom.get_coord(), distance_cutoff))
                    
                    # Get the residues of these atoms (excluding the hetero residue itself)
                    for atom in nearby_atoms:
                        res = atom.get_parent()
                        res_id = res.get_id()
                        # Only include protein residues (not hetero, not water)
                        if res_id[0] == " " and res != residue:
                            res_info = f"{res.get_resname()} (Chain {res.get_parent().id}, Res #{res_id[1]})"
                            if res_info not in nearby_residues[hetero_key]:
                                nearby_residues[hetero_key].append(res_info)
    
    return nearby_residues

# Example usage:
# pdb_file = "path/to/your/pdb_file.pdb"
# hetero_residues = ["ATP", "MG", "NAG"]
# results = find_residues_near_hetero(pdb_file, hetero_residues)
# 
# for hetero, nearby in results.items():
#     print(f"\nResidues within 5Ã… of {hetero}:")
#     for res in nearby:
#         print(f"  {res}")


CASP15_substrate_dict = {
    "T1112": ["MPO"],
    "T1124": ["SAH"],
    "T1127": ["COA"],
    "T1146": ["NAG"],
    "T1170": ["ADP", "AGS"],
    "T1188": ["DW0"],
    "H1114": ["3NI", "F3S", "FCO", "MQ7"],
    "H1185": ["ADP", "ATP"],
}
CASP15_molecule_nearby_residues = {}
for pdb_file in pdb_files:
    pdb_name = os.path.basename(pdb_file)
    submit_target = pdb_name[:5]
    if submit_target in CASP15_substrate_dict:
        results = find_residues_near_hetero(pdb_file, CASP15_substrate_dict[submit_target])
        print(pdb_name, ":",results)
        CASP15_molecule_nearby_residues[pdb_name] = results


CASP15_metal_dict = {
    "T1112": ["NA"],
    "T1122": ["NA"],
    "T1145": ["CA"],
    "T1170": ["MG"],
    "T1177": ["MG"],
    "T1181": ["CA", "ZN"],
    "T1183": ["MG"],
    "T1187": ["ZN"],
    "T1188": ["CD", "CO"],
    "T1192": ["MG"],
    "H1114": ["MG"],
    "H1185": ["MG"],
}
CASP15_metal_nearby_residues = {}
for pdb_file in pdb_files:
    pdb_name = os.path.basename(pdb_file)
    submit_target = pdb_name[:5]
    if submit_target in CASP15_metal_dict:
        results = find_residues_near_hetero(pdb_file, CASP15_metal_dict[submit_target])
        print(pdb_name, ":",results)
        CASP15_metal_nearby_residues[pdb_name] = results
```

```python
def match_submit_to_processed(submit_target, processed_targets):
    def three_to_one(resname):
        """Convert 3-letter amino acid code to 1-letter."""
        from Bio.Data.IUPACData import protein_letters_3to1
        return protein_letters_3to1[resname.capitalize()]
    
    submit_parser = PDBParser(QUIET=True)
    submit_structure = submit_parser.get_structure("submit", submit_target)
    submit_chains = {}
    submit_residue_ids = {}
    submit_model_count = 0
    for model in submit_structure:
        submit_model_count += 1
        for chain in model:
            chain_id = chain.id
            residues = []
            residue_ids = []
            for residue in chain:
                if residue.get_id()[0] == " ":
                    resname = residue.get_resname()
                    try:
                        residues.append(three_to_one(resname))
                    except:
                        residues.append("X")
                    residue_ids.append(residue.get_id())
            seq = "".join(residues)
            assert len(seq) == len(residue_ids)
            submit_chains[chain_id] = seq
            submit_residue_ids[chain_id] = residue_ids
    if submit_model_count > 1:
        print("WARNING:", submit_target, f"has {submit_model_count} models")

    processed_chains = {}
    processed_residue_ids = {}
    for processed_target in processed_targets:
        processed_parser = PDBParser(QUIET=True)
        processed_structure = processed_parser.get_structure(
            f"processed_{processed_target.split('/')[-1]}", processed_target)
        for model in processed_structure:
            for chain in model:
                chain_id = chain.id
                if chain_id == " ":
                    chain_id = "A"
                residues = []
                residue_ids = []
                for residue in chain:
                    if residue.get_id()[0] == " ":
                        resname = residue.get_resname()
                        try:
                            residues.append(three_to_one(resname))
                        except:
                            residues.append("X")
                        residue_ids.append(residue.get_id())
                seq = "".join(residues)
                assert len(seq) == len(residue_ids)
                processed_chains[processed_target.split("/")[-1]] = seq
                processed_residue_ids[processed_target.split("/")[-1]] = residue_ids
    
    # print(submit_target.split("/")[-1],submit_chains)
    # print(submit_target.split("/")[-1],submit_residue_ids)
    # print(processed_chains)
    # print(processed_residue_ids)


    # now for each chain in submit, find the chain with highest similarity using sequence alignment
    from Bio import pairwise2
    from Bio.pairwise2 import format_alignment
    
    chain_mapping = {}

    for submit_chain_id, submit_seq in submit_chains.items():
        best_score = -float('inf')
        best_chain = None
        best_seq = None
        for processed_chain_id, processed_seq in processed_chains.items():
            # print(submit_chain_id, processed_chain_id)
            alignments = pairwise2.align.globalxx(submit_seq, processed_seq)
            if alignments:
                score = alignments[0].score
                # print(alignments[0])
                # I also need to get the aligned positions to residue number mapping
                # aligned_seq1, aligned_seq2, score, begin, end = alignments[0]
                # print("aln1", aligned_seq1)
                # print("aln2", aligned_seq2)
                similarity = score / len(submit_seq)
                if similarity > best_score:
                    best_score = similarity
                    best_chain = processed_chain_id
                    best_seq = processed_seq
                    # offset_by_align = 0
                    # if aligned_seq1.startswith('-') and not aligned_seq2.startswith('-'):
                    #     offset_by_align = aligned_seq1.index(aligned_seq1.strip('-')[0])
                    # elif aligned_seq2.startswith('-') and not aligned_seq1.startswith('-'):
                    #     offset_by_align = -aligned_seq2.index(aligned_seq2.strip('-')[0])
                    # else:
                    #     offset_by_align = 0
        # try:
        #     # print(processed_residue_ids[best_chain][0][1], submit_residue_ids[submit_chain_id][0][1], offset_by_align)
        #     offset = processed_residue_ids[best_chain][0][1] - submit_residue_ids[submit_chain_id][0][1] + offset_by_align
        # except:
        #     offset = 0
        

        if best_score > 0.8:
            mapping = {}
            best_alignments = pairwise2.align.globalxx(submit_seq, best_seq)
            best_alignment = best_alignments[0]
            aligned_seq1, aligned_seq2, score, begin, end = best_alignment
            print(aligned_seq1)
            print(aligned_seq2)
            print(submit_residue_ids[submit_chain_id])
            print(processed_residue_ids[best_chain])
            # print(submit_target, best_chain)
            seq_1_gap_number = aligned_seq1.count('-')
            seq_2_gap_number = aligned_seq2.count('-')
            # print(len(aligned_seq1), len(aligned_seq2), len(submit_residue_ids[submit_chain_id]), len(processed_residue_ids[best_chain]))
            # make sure the length of aligned sequences are the same as the length of the original sequences

            assert len(aligned_seq1) == len(aligned_seq2) == len(submit_residue_ids[submit_chain_id])+seq_1_gap_number == len(processed_residue_ids[best_chain])+seq_2_gap_number
            # now start to do residue number mapping
            # track positions in original sequences (skipping gaps)
            seq1_pos = 0  # position in submit sequence
            seq2_pos = 0  # position in processed sequence
            
            # iterate through aligned sequences
            for i in range(len(aligned_seq1)):
                # if neither position is a gap, map the residues
                if aligned_seq1[i] != '-' and aligned_seq2[i] != '-':
                    submit_res = submit_residue_ids[submit_chain_id][seq1_pos][1]
                    proc_res = processed_residue_ids[best_chain][seq2_pos][1] 
                    mapping[submit_res] = proc_res
                    seq1_pos += 1
                    seq2_pos += 1
                # handle gaps in either sequence
                elif aligned_seq1[i] == '-':
                    seq2_pos += 1
                else: # aligned_seq2[i] == '-'
                    seq1_pos += 1
            chain_mapping[submit_chain_id] = [best_chain, best_score, mapping]
    return chain_mapping

CASP15_chain_mapping = {}
for submit_target in submit_targets_list:
    ID = submit_target[1:5]
    # find any processed target with 1:5 same as ID
    processed_targets = [target for target in processed_targets_list if target[1:5] == ID]
    if len(processed_targets) == 0:
        print(submit_target, "has no processed target")
    else:
        chain_mapping = match_submit_to_processed(os.path.join(CASP15_submit_path, submit_target),
                                [ os.path.join(CASP15_processed_path, processed_target) for processed_target in processed_targets])
        CASP15_chain_mapping[submit_target] = chain_mapping
CASP15_chain_mapping
```

```python
CASP15_target_nearby_residues_filtered = {}
# do a double check
for target_chain_id, residues in CASP15_target_nearby_residues.items():
    target_file = os.path.join(CASP15_processed_path, target_chain_id)
    with open(target_file, "r") as f:
        for line in f:
            if line.startswith("ATOM"):
                res_num = int(line[22:26].strip())
                res_name = line[17:20].strip()
                res_name_1 = three_to_one(res_name)
                for residue in residues:
                    # print(type(res_num), type(residue[0]))
                    if res_num == residue[0]:
                        if res_name_1 == residue[1]:
                            if target_chain_id not in CASP15_target_nearby_residues_filtered:
                                CASP15_target_nearby_residues_filtered[target_chain_id] = []
                            if (res_num, res_name_1) not in CASP15_target_nearby_residues_filtered[target_chain_id]:
                                CASP15_target_nearby_residues_filtered[target_chain_id].append((res_num, res_name_1))
                        else:
                            print("WARNING:",target_chain_id, res_num, res_name_1, residue[1])

CASP15_target_nearby_residues_filtered
```
